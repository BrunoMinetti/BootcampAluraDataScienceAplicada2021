```
def baixar_arquivo(url, endereco):
    resposta = req.get(url)
    if resposta.status_code == req.codes.OK:
        with open(endereco, 'wb') as novo_arquivo:
                novo_arquivo.write(resposta.content)
        print("Download finalizado. Arquivo salvo em: {}".format(endereco))
    else:
        resposta.raise_for_status()
```

```
def plot_bar_race_covid(dados, titulo, quantidade_dias = None):
    if quantidade_dias is not None:
        dados = dados.tail(quantidade_dias)
    return bcr.bar_chart_race(df = dados,
        orientation='h',
        sort='desc',
        n_bars=27,
        fixed_order=False,
        fixed_max=False,
        steps_per_period=10,
        interpolate_period=False,
        label_bars=True,
        bar_size=.95,
        period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
        period_fmt='%d/%m/%Y',
        period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                        's': f'Total de {titulo}: {v.nlargest(27).sum():,.0f}',
                                        'ha': 'right', 'size': 8, 'family': 'Courier New'},
        perpendicular_bar_func='median',
        period_length=225,
        figsize=(5, 3),
        dpi=144,
        cmap='dark12',
        title=f'Quantidade de {titulo} por estados - {data}',
        title_size='',
        bar_label_size=7,
        tick_label_size=7,
        shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
        scale='linear',
        writer=None,
        fig=None,
        bar_kwargs={'alpha': .7},
        filter_column_colors=False)
```

```
def plot_media_movel_covid_nacional_indicador (indicador, data):
    dia = data
    df = df_covid.query('regiao == "Brasil" and populacaoTCU2019.notnull()\
     and municipio.isnull()', engine='python')[['data', indicador]]
    df[f'mediaMovel{indicador}'] = df[indicador].rolling(14).mean()
    trace1 = go.Bar(x = df['data'],
                    y = df[indicador],
                    name=indicador,
                    marker = {'color': '#2125c4'})
    trace2 = go.Scatter(x =df['data'],
                y = df[f'mediaMovel{indicador}'],
                mode = 'lines',
                name='Média Móvel de '+indicador,
                line = {'color': '#FF2F0A'})
    data = [trace1,trace2]
    layout = go.Layout(title = f"Indicadores de {indicador} Brasil com a média móvel em {dia}",
                    xaxis = {'title': 'Data'},
                    yaxis = {'title': indicador},
                    barmode = 'stack',     legend=dict(x=0,y=1.0))
    fig = go.Figure(data=data, layout=layout)  
    py.iplot(fig)
```

```
def plot_covid_nacional_indicador (indicador, data):
    dia = data
    fig=go.Figure()
    df = df_covid.query('regiao == "Brasil" and municipio.isnull() and\
    populacaoTCU2019.notnull()',engine='python')[['data', indicador]]
    trace = go.Scatter(x = df['data'],
                    y = df[indicador],
                    mode = 'lines',
                    marker = {'color': '#ed0239'})
    data = [trace]
    layout = go.Layout(title=f"Indicadores de {indicador} no Brasil em {dia}",
                    yaxis={'title':indicador},
                    xaxis={'title': 'Data'})
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig)
```

## Tratamento dos dados

```
#Estratégia de preenchimento da tabela propagando os dados que já tinhamos.
#Foi discutido com o pessoal do Sírio Libanês
def preenche_tabela(dados):
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True,axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

Por Recomendação do Hospital, vamos desconsiderar os pacientes que foram para a UTI na primeira janela.

```
dados_limpos = preenche_tabela(df)
# Remover os dados da primeira janela, por orientação do Sírio Libanês, visto
# que não é interessante as pessoas que chegaram e já foram para a UTI
a_remover = dados_limpos.query("WINDOW=='0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query("PATIENT_VISIT_IDENTIFIER not in @a_remover")
#Mesmo com essa limpeza ainda ficou uma linha não preenchida, estamos dropando
dados_limpos = dados_limpos.dropna()
dados_limpos.describe()
```

Nessa parte vamos tentar pegar apenas os dados da primeira janela (0-2) dos pacientes que deram entrada na UTI.

```
# Essa função resolve a questão de termos mais de uma linha para cada paciente,
# pois o dado está organizado em janelas categóricas de período.
def prepare_window(rows):
    if(np.any(rows["ICU"])):
        rows.loc[rows["WINDOW"]=="0-2", "ICU"] = 1
    return rows.loc[rows["WINDOW"] == "0-2"]

dados_limpos = dados_limpos.groupby("PATIENT_VISIT_IDENTIFIER").apply(prepare_window)
# Transformando dados categóricos "AGE_PERCETIL" de string para inteiro
dados_limpos.AGE_PERCENTIL = dados_limpos.AGE_PERCENTIL.astype("category").cat.codes
dados_limpos.head()
```

## Testando Modelos Preditivos:
Estudo preliminar com todas as variáveis, só para tentar verificar padrões e estudar as métricas que usaremos, quando formos escolher o melhor modelo posteriormente, selecionando as variáveis ideais para aplicar no nosso modelo.

### Dummy Classifier

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

np.random.seed(458963)

# Criando um modelo Dummy Classifier
x_columns = df.columns
y = dados_limpos["ICU"]
x = dados_limpos[x_columns].drop(["ICU","WINDOW"], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y)
modelo = DummyClassifier()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

#### Matriz de Confusão:





```
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plot_confusion_matrix(modelo, x_test, y_test, cmap=plt.cm.Greens_r, ax=ax)
plt.title('Matriz de Confusão - Dummy Classifier');
```

#### Relatório de Classificação:
- Precisão - todas as classificações positivas, quantas são realmente positivas.
- Recall - todas as classificações positivas, quantas eu realmente classifiquei corretamente

```
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```
